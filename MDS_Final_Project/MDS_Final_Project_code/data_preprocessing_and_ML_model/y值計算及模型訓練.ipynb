```python
# 定義計算U的函數
def calculate_average(df, column_name, weight_column,new_column_name):
    df[new_column_name] = df.groupby("橋梁 系統ID")[column_name].transform("mean")
    # Calculate the weighted sum and sum of weights for each group
    weighted_sum = df.groupby("橋梁 系統ID").apply(lambda group: (group[weight_column] * group[column_name]).sum())
    sum_weights = df.groupby("橋梁 系統ID")[weight_column].transform("sum")
    
    # Calculate the weighted average and assign it to the new column
    df[new_column_name] = weighted_sum / sum_weights
    
    return df
# 遍歷六個DataFrame進行處理
for year, df in zip(range(2018, 2024), [df2018, df2019, df2020, df2021, df2022, df2023]):
    # 計算平均值並新增相應的列，要計算檢測前和檢測後得所以要執行兩次
    df = calculate_average(df, "U", "權重", "U_mean")
    df = calculate_average(df, "U.1","權重", "U_mean_after")

    # 選擇需要的列
    df_selected = df[["橋梁 系統ID", "U_mean", "U_mean_after", "檢測日期", "橋梁名稱"]]

    # 刪除重複的行
    df_selected = df_selected.drop_duplicates()

    # 重設索引
    df_selected = df_selected.reset_index(drop=True)

    # 將結果存儲到新的DataFrame
    globals()[f"df{year}_U"] = df_selected
    print(f"{year}年檢測橋梁數: {df_selected.shape[0]}")
```

```python
import pandas as pd

# 假設你有 df2018_U, df2019_U, df2020_U, ..., df2023_U

# 將這些 DataFrame 存放到一個列表中
dfs_list = [df2018_U, df2019_U, df2020_U, df2021_U, df2022_U, df2023_U]

# 創建一個字典來存放每年的平均值
average_dict = {}

# 遍歷每個 DataFrame，計算 U_mean 和 U_mean_after 的平均值
for year, df in zip(range(2018, 2024), dfs_list):
    U_mean_avg = df['U_mean'].mean()
    U_mean_after_avg = df['U_mean_after'].mean()
    
    average_dict[f"{year}_U_mean"] = U_mean_avg
    average_dict[f"{year}_U_mean_after"] = U_mean_after_avg

# 將結果轉換成 DataFrame
average_df = pd.DataFrame(list(average_dict.items()), columns=['Column', 'Average'])

# 輸出結果
print(average_df)
```

```python
import pandas as pd

# 創建一個包含所有六個DataFrame的列表
dfs_list = [df2018_U, df2019_U, df2020_U, df2021_U, df2022_U, df2023_U]

# 使用pd.concat合併DataFrame
merged_df = pd.concat(dfs_list, ignore_index=True)

# 根據橋梁系統ID和檢查日期排序DataFrame
merged_df = merged_df.sort_values(by=["橋梁 系統ID", "檢測日期"], ascending=[True, True])

# 將橋梁名稱移至第一列
merged_df = merged_df[["橋梁名稱"] + [col for col in merged_df.columns if col != "橋梁名稱"]]

# 將檢測日期轉換成"YYYY-MM-DD"的格式
merged_df["檢測日期"] = pd.to_datetime(merged_df["檢測日期"]).dt.strftime("%Y-%m-%d")

# 將合併後的DataFrame輸出到Excel檔案
output_file = "橋梁歷年劣化程度.xlsx"
#merged_df.to_excel(output_file, index=False)
merged_df['U_mean_all'] = merged_df.groupby("橋梁 系統ID")['U_mean'].transform('mean')
```

計算劣化速度平均

```python

# Convert '檢測日期' to datetime format
merged_df['檢測日期'] = pd.to_datetime(merged_df['檢測日期'], errors='coerce')

# Convert columns to numeric types
merged_df['U_mean'] = pd.to_numeric(merged_df['U_mean'], errors='coerce')
merged_df['U_mean_after'] = pd.to_numeric(merged_df['U_mean_after'], errors='coerce')

# Calculate U 的平均變化速度
# 計算 U_change
# Calculate U_change
u_mean_after_shifted = merged_df.groupby("橋梁 系統ID")['U_mean_after'].shift(1)
date_difference = ((merged_df['檢測日期'] - merged_df.groupby("橋梁 系統ID")['檢測日期'].shift(1)).dt.days)/30
merged_df['U_change'] = (merged_df['U_mean'] - u_mean_after_shifted) / date_difference
merged_df.dropna(subset=['U_change'], inplace=True)
merged_df['U_change_ave'] = merged_df.groupby("橋梁 系統ID")['U_change'].transform('mean')
merged_df
```
